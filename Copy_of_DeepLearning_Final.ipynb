# 1.Introduction
Traffic sign recognition is a multi-category classification problem with unbalanced class frequencies. It is a challenging real-world computer vision problem of high practical relevance, which has been a research topic for several decades. Many studies have been published on this subject and multiple systems, which often restrict themselves to a subset of relevant signs, are already commercially available in new high- and mid-range vehicles. Nevertheless, there has been little systematic unbiased comparison of approaches and comprehensive benchmark datasets are not publicly available. Road signs are designed to be easily detected and recognized by human drivers. They follow clear design principles using color, shape, icons and text. These allow for a wide range of variations between classes. Signs with the same general meaning, such as the various speed limits, have a common general appearance, leading to subsets of traffic signs that are very similar to each other. Illumination changes, partial occlusions, rotations, and weather conditions further increase the range of variations in visual appearance a classifier has to cope with. Humans are capable of recognizing the large variety of existing road signs in most situations with near-perfect accuracy

```python
import os
import numpy as np
from sklearn.model_selection import train_test_split
import shutil

# Define the path to the .ppm data
root_dir = r"/content/GTSRB/Training"
training_dir = os.path.join(os.path.dirname(root_dir), 'Training_70')
testing_dir = os.path.join(os.path.dirname(root_dir), 'Testing_10')
validation_dir = os.path.join(os.path.dirname(root_dir), 'Validation_20')

dir_list = os.listdir(root_dir)

for dir_name in dir_list:
    if os.path.isdir(os.path.join(root_dir, dir_name)):
        print(f"Reading file names in {dir_name}...")
        file_list = [f for f in os.listdir(os.path.join(root_dir, dir_name)) if f.endswith('.ppm')]
        file_path_list = [os.path.join(root_dir, dir_name, f) for f in file_list]
        if len(file_path_list) > 0:
            train_file_path_list, remaining_file_path_list = train_test_split(file_path_list, train_size=0.7, random_state=42)
            test_file_path_list, validation_file_path_list = train_test_split(remaining_file_path_list, train_size=0.333, random_state=42)

            dest_train_dir = os.path.join(training_dir, dir_name)
            dest_val_dir = os.path.join(validation_dir, dir_name)
            dest_test_dir = os.path.join(testing_dir, dir_name)

            if not os.path.exists(dest_train_dir):
                os.makedirs(dest_train_dir)
            if not os.path.exists(dest_val_dir):
                os.makedirs(dest_val_dir)
            if not os.path.exists(dest_test_dir):
                os.makedirs(dest_test_dir)

            for file_path in train_file_path_list:
                file_name = os.path.basename(file_path)
                dest = os.path.join(dest_train_dir, file_name)
                if os.path.exists(file_path):
                    shutil.move(file_path, dest)
                    print(f"Moving {file_name} to {dest_train_dir}...")
                else:
                    print(f"Skipping {file_name}. File does not exist.")

            for file_path in validation_file_path_list:
                file_name = os.path.basename(file_path)
                dest = os.path.join(dest_val_dir, file_name)
                if os.path.exists(file_path):
                    shutil.move(file_path, dest)
                    print(f"Moving {file_name} to {dest_val_dir}...")
                else:
                    print(f"Skipping {file_name}. File does not exist.")

            for file_path in test_file_path_list:
                file_name = os.path.basename(file_path)
                dest = os.path.join(dest_test_dir, file_name)
                if os.path.exists(file_path):
                    shutil.move(file_path, dest)
                    print(f"Moving {file_name} to {dest_test_dir}...")
                else:
                    print(f"Skipping {file_name}. File does not exist.")
        else:
            print(f"No ppm files found in {dir_name}...")
```

# 3 Transfer Learning

```python
import os
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define the path to the training dataset
train_dir = "/content/GTSRB/Training"

# Define the data augmentation parameters
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

# Define the batch size
batch_size = 32

# Create data generators for training sets
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Path for saving augmented images
augment_dir = r"augmen_data"

# Ensure augmented_data directory exists
if not os.path.exists(augment_dir):
    os.makedirs(augment_dir)

# Get the list of class names and ensure directories exist for each class
classes = list(train_generator.class_indices.keys())
for class_name in classes:
    class_dir = os.path.join(augment_dir, class_name)
    if not os.path.exists(class_dir):
        os.makedirs(class_dir)

# Set the target number of images per class
target_images_per_class = 40 # Set this to the desired number of images per class

# Initialize a dictionary to track the number of images per class
class_counts = {class_name: 0 for class_name in classes}

# Loop to generate images until each class has the target number of images
i = 0
while True:
    # Generate a batch of images
    X_batch, y_batch = next(train_generator)

    for img, label in zip(X_batch, y_batch):
        img = (img * 255).astype('uint8')
        class_index = np.argmax(label)
        class_name = classes[class_index]

        # Save only if the class hasn't reached the target number of images
        if class_counts[class_name] < target_images_per_class:
            cv2.imwrite(os.path.join(augment_dir, class_name, f'img_{i}.jpg'), img)
            class_counts[class_name] += 1
            i += 1

        # Check if all classes have reached the target number of images
        if all(count >= target_images_per_class for count in class_counts.values()):
            print("All classes have reached the target number of images.")
            break

    # Break the outer loop if all classes have enough images
    if all(count >= target_images_per_class for count in class_counts.values()):
        break

# Print the number of images generated for each class
for class_name, count in class_counts.items():
    print(f"Class {class_name}: {count} images")
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Flatten, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.regularizers import l2
from keras.callbacks import EarlyStopping, LearningRateScheduler

os.environ['TF_ENABLE_ONEDNN_OPTS'] = '0'

# Define the path to the training dataset
train_dir = "/content/augmen_data"
validation_dir = "/content/GTSRB/Testing"

# Define the data augmentation parameters
datagen = ImageDataGenerator(
    rescale=1./255,
)

# Define the batch size
batch_size = 64

# Create data generators for training and validation sets
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

validation_generator = datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

# Load the pre-trained MobileNetV2 model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Freeze all layers except the last 10 layers
for i, layer in enumerate(base_model.layers):
    if i < len(base_model.layers) - 10:
        layer.trainable = False

# Add a new classification head
x = base_model.output
x = Flatten()(x)
x = BatchNormalization()(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dense(43, activation='softmax')(x)

# Create the new model
model = Model(inputs=base_model.input, outputs=x)

# Compile the model
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

# Implement a learning rate schedule
def schedule(epoch, lr):
    if epoch < 5:
        return 0.001
    else:
        return 0.0001

lr_scheduler = LearningRateScheduler(schedule)

# Use early stopping
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5, min_delta=0.001)

# Train the model
history = model.fit(train_generator, validation_data=validation_generator, epochs=20, callbacks=[lr_scheduler, early_stopping])

# Plot the training and validation accuracy and loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy')
plt.legend()
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss')
plt.legend()
plt.show()
```
